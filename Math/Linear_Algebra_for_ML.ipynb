Question :
A bank robber start with 150 km/h speed. A police start after 5 min late with 180 speed. When the police catch the robber?

Solution :
robber speed = 150 km/h = 2.5 km/min
police speed = 180 km/h = 3 km/min
E1 for robber : d = 2.5t           [d = vt]
E2 for police : d = 3 (t-5)

Now we can say,
2.5t = 3(t-5)
=> t = 30 min

And also,
d = 2.5 * 30 = 75 km

```python
speed_robber = 150 / 60  # Convert km/h to km/min
speed_police = 180 / 60  # Convert km/h to km/min
delay_police = 5  # Delay in minutes

# Compute intersection time
t_intersect = (speed_police * delay_police) / (speed_police - speed_robber)
d_intersect = speed_robber * t_intersect
```

```python

t = np.linspace(0,50,1000)
d_robber = 2.5*t
d_police = 3*(t-5)
```

```python
plt.title("Catch the robber")
plt.xlabel("Time (min)")
plt.ylabel("Distance (km)")

plt.xlim(0,50)
plt.ylim(0,100)
plt.xticks(np.arange(0,50,5))
plt.yticks(np.arange(0,100,5))
plt.plot(t,d_robber,c = 'r',label = "Robber")
plt.plot(t,d_police,c = 'g',label = "Police")
plt.scatter(t_intersect, d_intersect, c = 'purple', label = "Answer point")
plt.axvline(x = t_intersect, c = 'purple', linestyle = '--')
plt.axhline(y = d_intersect, c = 'purple', linestyle = '--')
plt.legend()
```

In the context of mathematics, physics, and machine learning, a **tensor** is a generalization of scalars, vectors, and matrices to higher dimensions. Tensors are data structures that represent multidimensional arrays and are used extensively in fields like deep learning, scientific computing, and computer graphics.

---

**Definition**
A tensor is a mathematical object that can be described as an array of numbers arranged on a regular grid with a specific number of dimensions. It is the building block of data in machine learning frameworks like TensorFlow and PyTorch.

---

**Types of Tensors**
The dimensionality of a tensor determines its type:
1. **Scalar (0-dimensional tensor):**
    - A single number.
    - Example: 5 , -3.14 .
    - Shape: `()` (no dimensions).

2. **Vector (1-dimensional tensor):**
    - A 1D array of numbers.
    - Example: [1, 2, 3].
    - Shape: `(n)` (where n is the number of elements).

3. **Matrix (2-dimensional tensor):**
    - A 2D array of numbers.
    - Example:
      [[1, 2],
      [3, 4]]
    - Shape: `(m, n)` (where m is the number of rows and n is the number of columns).

4. **Higher-Dimensional Tensors (3D and beyond):**
    - Multi-dimensional arrays.
    - Example: A 3D tensor can represent RGB image data (height, width, color channels).
    - Shape: `(d_1, d_2, ..., d_n)`.

---

**Why are Tensors Important?**
1. **Representation of Data:**
    - Tensors efficiently store and manipulate multi-dimensional data, such as images, videos, and text embeddings.
    
2. **Mathematical Operations:**
    - Many mathematical operations (e.g., dot products, matrix multiplications) can be generalized to tensors.
    
3. **Deep Learning:**
    - Neural networks use tensors to store inputs, weights, activations, and gradients.
    - Frameworks like TensorFlow and PyTorch rely heavily on tensors for computation.

4. **Physics:**
    - Tensors are used to describe physical quantities like stress, strain, and electromagnetic fields, which may have components in multiple dimensions.

Function for Plot

```python
def plot_vectors(vectors, colors):
    """
    Plot one or more vectors in a 2D plane, specifying a color for each.

    Arguments
    ---------
    vectors: list of lists or of arrays
        Coordinates of the vectors to plot. For example, [[1, 3], [2, 2]]
        contains two vectors to plot, [1, 3] and [2, 2].
    colors: list
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.

    Example
    -------
    plot_vectors([[1, 3], [2, 2]], ['red', 'blue'])
    plt.xlim(-1, 4)
    plt.ylim(-1, 4)
    """
    plt.figure()
    plt.axvline(x=0, color='lightgray')
    plt.axhline(y=0, color='lightgray')

    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],angles='xy', scale_units='xy', scale=1, color=colors[i],)
```

```python
img_arr = np.asarray(list(img.getdata(band=0)),float)
imgmat = np.reshape(img_arr,(img.size[1],img.size[0]))
imgmat = np.matrix(imgmat)
```

```python
u, s, vh = np.linalg.svd(imgmat)

compression_levels = [2, 4, 8, 16, 32, 64, 128 , 256 , 512]
for i in compression_levels:
    res = np.dot(u[:, :i], np.diag(s[:i]) @ vh[:i, :])
    plt.imshow(res, cmap='gray')
    plt.title(f"Compressed with {i} Singular Values")
    plt.show()
```

Moore-Penrose Pseudoinverse Solve the unknown values

```python
x1 = [0, 1, 2, 3, 4, 5, 6, 7.]
y = [1.86, 1.31, .62, .33, .09, -.67, -1.23, -1.37]
```

```python
plt.title('Clinical Trial')
plt.xlabel('Drug dosage (ml)')
plt.ylabel('Forgetfulness')
plt.scatter(x1,y)
```

```python
x0 = np.ones(8)
x0
```

```python
x = np.concatenate((np.matrix(x0).T, np.matrix(x1).T), axis = 1)
x
```

The weights w = xy;  where x is the pseudoinverse of matrix x.

```python
w = np.dot(np.linalg.pinv(x),y)
w
```

```python
a = np.asarray(w).reshape(-1)[0]
b = np.asarray(w).reshape(-1)[1]
```

```python
fig, ax = plt.subplots()

plt.title('Clinical Trial')
plt.xlabel('Drug dosage (ml)')
plt.ylabel('Forgetfulness')

ax.scatter(x1, y)

x_min, x_max = ax.get_xlim()
y_at_xmin = b*x_min + a
y_at_xmax = b*x_max + a

ax.set_xlim([x_min, x_max])
_ = ax.plot([x_min, x_max], [y_at_xmin, y_at_xmax], c='C01')
```
